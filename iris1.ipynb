```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
from sklearn import datasets
```

```python
data=datasets.load_iris()
```

```python
diris=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
tiris=pd.DataFrame(data.target,columns=["target"])
```

```python
iris=pd.concat([diris,tiris],axis=1)
```

```python
iris
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
lr=LogisticRegression()
```

```python
rfc=RandomForestClassifier()
```

```python
dtc=DecisionTreeClassifier()
```

```python
iris.columns
```

```python
X=iris.drop('target',axis=1)
y=iris['target']
```

```python
lr.fit(X,y)
```

```python
rfc.fit(X,y)
```

```python
dtc.fit(X,y)
```

```python
plr=lr.predict(X)
```

```python
prfc=rfc.predict(X)
```

```python
pdtc=dtc.predict(X)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y,plr)
```

```python
accuracy_score(y,prfc)
```

```python
accuracy_score(y,pdtc) #decision tree classifier
```
